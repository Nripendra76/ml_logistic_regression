```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, StratifiedKFold

model = LogisticRegression(max_iter=1000)

param_grid = [
    {
        'penalty': ['l1'],
        'solver': ['liblinear', 'saga'],
        'C': [100,10,1.0,0.1,0.01]
    },
    {
        'penalty': ['l2'],
        'solver': ['lbfgs','newton-cg','sag','saga','liblinear'],
        'C': [100,10,1.0,0.1,0.01]
    },
    {
        'penalty': ['elasticnet'],
        'solver': ['saga'],
        'C': [100,10,1.0,0.1,0.01],
        'l1_ratio': [0.5]
    }
]

cv = StratifiedKFold(n_splits=5)

grid = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='accuracy',
    cv=cv,
    n_jobs=-1
)

y_train = y_train.ravel()

grid.fit(X_train, y_train)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)

cm=confusion_matrix(y_test,y_pred)

cr=classification_report(y_test,y_pred)
print(score)
print(cm)
print(cr)
```

### logistic regression for imbalanced dataset

```python
from collections import Counter
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=10000,n_features=2,n_clusters_per_class=1,n_redundant=0,
                        weights=[0.99],random_state=10)
```

```python
Counter(y)
```

```python
# import seaborn as sns
# import pandas as pd
# sns.scatterplot(pd.DataFrame(X)[0],pd.DataFrame(X)[1],hue=y)
```

```python
df = pd.DataFrame(X)
df['target'] = y

sns.scatterplot(data=df, x=0, y=1, hue='target')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
##### hyperparameter tuning
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton_cg','lbfgs','liblinear','sag','saga']
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv)
```

```python
grid.fit(X_train,y_train)
```

```python
print(grid)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score=accuracy_score(y_test,y_pred)

cm=confusion_matrix(y_test,y_pred)

cr=classification_report(y_test,y_pred)
print(score)
print(cm)
print(cr)
```

### Logistic Regression With ROC curve and ROC AUC score

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
X,y=make_classification(n_samples=1000, n_classes=2 ,random_state=0)
```

```python
# Created a dummy model with default 0 as output 
dummy_model_prob = [0 for _ in range(len(y_test))]
```

```python
## Lets Create Basic Logistic Model
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
## Prediction based on probability
model_prob=model.predict_proba(X_test)
```

```python
## Lets focus on the positive outcome
model_prob=model_prob[:,1]
```

```python
## Lets calulcate the scores
dummy_model_auc=roc_auc_score(y_test,dummy_model_prob)
model_auc=roc_auc_score(y_test,model_prob)
print(dummy_model_auc)
print(model_auc)
```

```python
## calculate ROC Curves
dummy_fpr, dummy_tpr, _ = roc_curve(y_test, dummy_model_prob)
model_fpr, model_tpr, thresholds = roc_curve(y_test, model_prob)
```

```python
from matplotlib import pyplot
```

```python
# plot the roc curve for the model
pyplot.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
pyplot.plot(model_fpr, model_tpr, marker='.', label='Logistic')
# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
# show the legend
pyplot.legend()
# show the plot
pyplot.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8,6))

# plot curves
ax.plot(dummy_fpr, dummy_tpr, linestyle='--', label='Dummy Model')
ax.plot(model_fpr, model_tpr, marker='.', label='Logistic')

# annotate only some points (important!)
for i in range(0, len(thresholds), max(1, len(thresholds)//10)):
    ax.annotate(str(np.round(thresholds[i], 2)),
                (model_fpr[i], model_tpr[i]))

# labels
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')

# legend
ax.legend()

plt.show()
```
